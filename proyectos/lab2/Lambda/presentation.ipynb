    <h1 style="font-size: 1.4rem; margin: 0; letter-spacing: -0.2px;">Comprender las relaciones entre variables</h1>
  </div>

  <p style="color: #333; margin-top: 0.65rem; margin-bottom: 0.9rem; font-size: 1rem;">
    Muchos conjuntos de datos contienen múltiples variables cuantitativas y la visualización es la principal herramienta
    para entender, rápidamente, cómo se relacionan entre sí.
  </p>

  <div style="margin-top: 1rem; padding-top: 0.2rem;">
    <p style="font-size: 1rem; color: #444; margin-top: 0.4rem;">
      Para <span style="font-weight:600; color:#1b3b8f;">dos variables</span>, se puede usar un <strong>diagrama de dispersión (scatterplot)</strong>.
    </p>
    <p style="font-size: 1rem; color: #444; margin-top: 0.4rem;">
      Para <span style="font-weight:600; color:#1b3b8f;">más de dos variables</span>, se pueden usar:
    </p>
    <ul style="padding-left: 1rem; margin: 0.6rem 0;">
      <li><strong>Gráficos de burbujas:</strong> añaden una dimensión através del tamaño de la burbuja.</li>
      <li><strong>Matrices de dispersión:</strong> pares de variables en una cuadrícula para ver relaciones bivariadas.</li>
      <li><strong>Correlogramas:</strong> resumen correlaciones entre muchas variables (útiles para una visión general).</li>
    </ul>
  </div>
</div>

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os


# Directorio para guardar los gráficos generados
os.makedirs("images", exist_ok=True)

df = pd.read_csv(
    "../../../datasets/alumnos/lambda-09-happiness_report_dataset/data/World Happiness Report 2019 dataset.csv"
)
df
```

```python
# Visualizar una sola variable y guardar la imagen
plt.figure()
sns.histplot(df["GDP per capita"], bins=20)
plt.title("Distribución del GDP per capita (2019)")
plt.xlabel("GDP per capita")
plt.ylabel("Número de países")
plt.tight_layout()
plt.savefig("images/hist_gdp_per_capita.png", dpi=600)
plt.show()
```

```python
# Visualizar dos variables con un scatter plot
# Esto es simple de entender
sns.scatterplot(x="GDP per capita", y="Score", data=df)
plt.title("Relación entre GDP per capita y Score")
plt.xlabel("GDP per capita")
plt.ylabel("Score")
plt.tight_layout()
plt.savefig("images/scatter_gdp_per_capita_score.png", dpi=600)
plt.show()
```

```python
# Visualizar tres variables agregando color
# Todavía es fácil de entender
sns.scatterplot(
    x="GDP per capita",
    y="Score",
    data=df,
    hue="Healthy life expectancy",
    palette="RdYlGn",
)
plt.title("Relación entre GDP per capita, Score y Healthy life expectancy")
plt.xlabel("GDP per capita")
plt.ylabel("Score")
plt.legend(bbox_to_anchor=(1.05, 1), title="Healthy life expectancy")
plt.savefig(
    "images/scatter_gdp_per_capita_score_healthy_life_expectancy.png",
    dpi=600,
    bbox_inches="tight",
)
plt.show()
```

```python
# Visualizar cuatro variables agregando tamaño
# Interpretar la información comienza a volverse confuso
scatter = sns.scatterplot(
    x="GDP per capita",
    y="Score",
    data=df,
    hue="Healthy life expectancy",
    palette="RdYlGn",
    size="Social support",
    sizes=(10, 200),
    alpha=0.8,
)
plt.title(
    "Relación entre GDP per capita, Score, Healthy life expectancy y Social support"
)
plt.xlabel("GDP per capita")
plt.ylabel("Score")
plt.legend(bbox_to_anchor=(1, 1))
plt.savefig(
    "images/scatter_gdp_per_capita_score_healthy_life_expectancy_social_support.png",
    dpi=600,
    bbox_inches="tight",
)
plt.show()
```

```python
# Matriz de dispersión todos contra todos
# Este gráfico muestra todas las relaciones entre dos variables
# Esto es demasiada información y dificulta la interpretación
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionamos solo las columnas numéricas
numeric_cols = [
    "Score",
    "GDP per capita",
    "Social support",
    "Healthy life expectancy",
    "Freedom to make life choices",
    "Generosity",
    "Perceptions of corruption",
]

sns.pairplot(df[numeric_cols], corner=True)
plt.savefig("images/pairplot.png", dpi=600)
plt.show()
```

  <p style="color: #333; margin-top: 0.65rem; margin-bottom: 0.9rem; font-size: 1rem;">
    La elección del gráfico es crucial: algunas visualizaciones pueden ser engañosas si no se interpretan con cuidado.
  </p>

  <div style="margin-top: 1rem; padding-top: 0.2rem;">
    <p style="font-size: 1rem; color: #444; margin-top: 0.4rem;">
      Ejemplos y por qué pueden causar problemas:
    </p>
    <ul style="padding-left: 1rem; margin: 0.6rem 0;">
      <li>
        <strong>Gráficos de burbujas</strong>: usan <span style="font-weight:600; color:#1b3b8f;">posición</span> y <span style="font-weight:600; color:#1b3b8f;">tamaño</span>
        para codificar información, el tamaño es menos preciso visualmente que la posición, lo que dificulta comparar magnitudes.
      </li>
      <li>
        <strong>Correlogramas</strong>: al abstraer la correlación y ocultar los puntos, pueden ocultar
        patrones como la <em>no linealidad</em> u <em>outliers</em>, conduciendo a conclusiones equivocadas.
      </li>
    </ul>
  </div>
</div>

```python
# ============================
# COMPARATIVA ENTRE GRÁFICOS
# ============================

fig = plt.figure(figsize=(20, 16))

# Definir los mismos ejes para todos los gráficos
x = df["GDP per capita"]
y = df["Score"]


# ===========================
# 1. GRÁFICO BUENO
# ===========================
ax1 = fig.add_subplot(2, 2, 1)

scatter = ax1.scatter(
    x,
    y,
    c=df["Healthy life expectancy"],
    cmap="RdYlGn",
    s=150,
    edgecolors="black",
    linewidth=0.5,
    alpha=0.8,
)

# Barra de colores con etiqueta apropiada
cbar = plt.colorbar(scatter, ax=ax1)
cbar.set_label("Expectativa de vida")

ax1.grid(True, alpha=0.4, linestyle="-", linewidth=0.5)

ax1.set_xlabel("PBI per cápita (USD)")
ax1.set_ylabel("Nivel de felicidad")
plt.title(
    "1. BUENO: Scatter plot visualmente claro",
    fontsize=14,
    weight="bold",
    color="green",
    pad=15,
)

# ===========================
# 2. GRÁFICO FEO
# ===========================
ax2 = fig.add_subplot(2, 2, 2)

# Colores sin cohesión
scatter = ax2.scatter(
    x,
    y,
    c=df["Healthy life expectancy"],
    cmap="rainbow",
    s=80,
    linewidth=2,
    alpha=0.6,
)

# Título y etiquetas con nombres "malos"
ax2.set_xlabel("pbi_per_capita", fontsize=10)
ax2.set_ylabel("nivel_de_felicidad", fontsize=10)
ax2.set_title("PBI vs Felicidad", fontsize=16, color="red", weight="bold")

cbar = plt.colorbar(scatter, ax=ax2)
cbar.set_label("exp_vida")

# Cuadrícula que distrae y es fea
ax2.grid(True, alpha=0.8, linestyle="--", linewidth=2, color="gray")
ax2.set_facecolor("#f0f0f0")

# Bordes gruesos
for spine in ax2.spines.values():
    spine.set_linewidth(3)
    spine.set_color("black")

plt.title(
    "2. FEO: Gráfico con información correcta pero mala estética",
    fontsize=14,
    weight="bold",
    color="orange",
    pad=15,
)

# ===========================
# 3. GRÁFICO MALO
# ===========================
ax3 = fig.add_subplot(2, 2, 3)

# Gráfico de burbujas donde el tamaño es difícil de interpretar
# Se usa un área proporcional al PBI. Esto es una mala práctica
# porque el ojo humano no percibe el área de los círculos de manera lineal,
# lo que puede llevar a interpretaciones erróneas sobre la magnitud de los datos.
bubble_sizes = (df["GDP per capita"] / df["GDP per capita"].max()) * 1000

scatter = ax3.scatter(
    x,
    y,
    s=bubble_sizes,
    c=df["GDP per capita"],
    cmap="coolwarm",
    alpha=0.6,
    edgecolors="black",
)

# Etiquetas confusas
ax3.set_xlabel("PBI per cápita")
ax3.set_ylabel("Nivel de felicidad")

# Confusión: el color representa lo mismo que el tamaño
cbar = plt.colorbar(scatter, ax=ax3)
cbar.set_label("PBI per cápita")

plt.title(
    "3. MALO: Gráfico de burbujas confuso y difícil de leer",
    fontsize=14,
    weight="bold",
    color="red",
    pad=15,
)

# ===========================
# 4. GRÁFICO INCORRECTO
# ===========================
ax4 = fig.add_subplot(2, 2, 4)
bubble_sizes = (df["GDP per capita"] / df["GDP per capita"].max()) * 1000

scatter = ax4.scatter(
    x,
    y,
    s=bubble_sizes,
    alpha=0.6,
    edgecolors="black",
)

cbar = plt.colorbar(scatter, ax=ax4)
cbar.set_label("PBI per cápita")

# Nos olvidamos los labels de los ejes
plt.title(
    "4. INCORRECTO: Gráfico de burbujas inentendible",
    fontsize=14,
    weight="bold",
    color="darkred",
    pad=15,
)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle(
    "Comparación de Visualizaciones del World Happiness Report (2019)",
    fontsize=18,
    weight="bold",
    y=0.98,
)

plt.savefig("images/scatterplot_comparison.png", dpi=600)
plt.show()
```

    <h1 style="font-size: 1.4rem; margin: 0; letter-spacing: -0.2px;">Encontrar un equilibrio al visualizar datos complejos</h1>
  </div>

  <p style="color: #333; margin-top: 0.65rem; margin-bottom: 0.9rem; font-size: 1rem;">
    Para datasets con muchas dimensiones, la <strong>reducción de dimensionalidad</strong> ayuda a mostrar patrones relevantes sin
    abrumar con ruido.
  </p>

  <div style="margin-top: 1rem; padding-top: 0.2rem;">
    <p style="font-size: 1rem; color: #444; margin-top: 0.4rem;">
      ¿Por qué funciona? Los datasets que tienen muchas características o campos frecuentemente contienen variables correlacionadas que transmiten
      información redundante, y reducir estas dimensiones permite resumir esa información.
    </p>
    <ul style="padding-left: 1.15rem; margin: 0.6rem 0;">
      <li>
        <strong>Análisis de Componentes Principales (PCA)</strong>: crea nuevas variables (componentes) que capturan las
        direcciones de mayor varianza. Se usa para <em>visualizar</em> patrones y para preprocesamiento antes de modelar.
      </li>
    </ul>
    <div style="background: linear-gradient(90deg, rgba(255, 214, 172, 1), rgba(255, 230, 161, 1)); border-left: 4px solid #ff830fff; padding: 0.6rem 0.3rem; margin: 0.7rem 0; border-radius: 6px; font-size: 1rem;">
      Siempre se debe buscar un equilibrio entre <em>interpretabilidad</em> (gráficos fáciles de explicar) y <em>fidelidad</em> (no perder patrones importantes).
    </div>
  </div>
</div>

    <h1 style="font-size: 1.4rem; margin: 0; letter-spacing: -0.2px;">Visualización de datos espaciales: el poder del contexto</h1>
  </div>

  <p style="color: #333; margin-top: 0.65rem; margin-bottom: 0.9rem; font-size: 1rem;">
    Los mapas permiten representar datos vinculados a ubicaciones específicas en el mundo real, como la distribución de poblaciones, la infraestructura de una ciudad, o los recursos naturales.
  </p>

  <p style="font-size: 1rem; color: #444; margin-top: 0.4rem;">
    Al mostrar los datos en su <span style="font-weight:600; color:#1b3b8f;">contexto geoespacial</span> adecuado, podemos identificar patrones, tendencias y relaciones
    que serían invisibles en una tabla o un gráfico convencional.
  </p>

  <div style="background: linear-gradient(90deg, rgba(255, 214, 172, 1), rgba(255, 230, 161, 1)); border-left: 4px solid #ff830fff; padding: 0.6rem 0.3rem; margin: 0.7rem 0; border-radius: 6px; font-size: 1rem;">
    Una representación geográfica no es solo un fondo decorativo, sino un marco esencial para comprender y contextualizar los datos.
  </div>

  <div style="margin-top: 1rem; padding-top: 0.2rem;">
    <h2 style="font-size: 1rem; margin-top:1rem;">Desafíos y proyecciones</h2>
    <p>
      A pesar de ser intuitivos, los mapas presentan desafíos de diseño. El principal es la <strong>proyección</strong>,
      el proceso de aplanar la superficie esférica de la Tierra en un plano 2D. 
    </p>
    <ul style="padding-left: 1.15rem; margin: 0.6rem 0;">
      <li>Cualquier proyección introduce <span>distorsiones</span> inevitables.</li>
      <li>No es posible preservar <em>ángulos</em> y <em>áreas</em> al mismo tiempo.</li>
      <li>Al crear un mapa, debemos decidir qué priorizar: la <strong>forma exacta</strong> de las regiones o su <strong>tamaño relativo</strong>.</li>
    </ul>
    <div style="background: linear-gradient(90deg, rgba(255, 214, 172, 1), rgba(255, 230, 161, 1)); border-left: 4px solid #ff830fff; padding: 0.6rem 0.3rem; margin: 0.7rem 0; border-radius: 6px; font-size: 1rem;">
      <em>No existe un mapa perfecto</em>, pero si mapas adecuados a los objetivos de análisis.
    </div>
  </div>

  <div style="margin-top: 1rem; padding-top: 0.2rem;">
    <h2 style="font-size:1.2rem; margin-top:1rem;">Tipos de mapas para mostrar datos cuantitativos</h2>
    <ul style="padding-left: 1.15rem; margin: 0.6rem 0;">
      <li>
        <strong>Mapas de coropletas:</strong>
        colorean áreas geográficas para mostrar cómo una cantidad varía entre regiones.
        <p style="font-size: 0.92rem; color: #444; margin-top: 0.4rem;">
          Útiles para comparar valores entre países, provincias o distritos.<br>
          Atención a la <strong>escala de colores</strong>: una escala discreta con rangos claros suele ser más efectiva que una continua.
        </p>
      </li>
      <li>
        <strong>Cartogramas:</strong>
        distorsionan el tamaño de las áreas para que sean proporcionales a una variable.
        <p style="font-size: 1rem; color: #444; margin-top: 0.4rem;">
          Ejemplo: un país se agranda o encoge según su población. Impactante para visualizar datos demográficos.
        </p>
      </li>
      <li>
        <strong>Mapas de puntos y símbolos proporcionales:</strong>
        colocan puntos o símbolos en coordenadas exactas, variando su tamaño/color según la magnitud.
        <p style="font-size: 1rem; color: #444; margin-top: 0.4rem;">
          Útiles para <em>densidad de eventos</em> (ej. ubicaciones de Airbnb) o variables puntuales (ej. centrales eléctricas).
        </p>
      </li>
    </ul>
    <div style="background: linear-gradient(90deg, rgba(255, 214, 172, 1), rgba(255, 230, 161, 1)); border-left: 4px solid #ff830fff; padding: 0.6rem 0.3rem; margin: 0.7rem 0; border-radius: 6px; font-size: 1rem;">
      Recomendación: antes de elegir el tipo de mapa, se debe pensar qué se quiere resaltar, ya sea distribución espacial, valores absolutos, proporciones, concentración de eventos, etcétera.
    </div>
  </div>
</div>

```python
import geopandas as gpd
from shapely.affinity import scale

# Dataset de geometrías
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# Juntar el dataset de geometrías con el de felicidad
world = world.rename(columns={"name": "Country"})
df = df.rename(columns={"Country or region": "Country"})
gdf = world.merge(df, on="Country", how="left")

# ===========================
# 1. GRÁFICO BUENO
# ===========================
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
gdf.plot(ax=ax, color="lightgrey")
mappable = gdf.plot(
    ax=ax, column="Score", cmap="RdYlGn", edgecolor="black", linewidth=0.25, legend=True
)

cbar = ax.get_figure().get_axes()[-1]
cbar.set_ylabel("Happiness Score", fontsize=12)

# Eliminar ticks de latitud y longitud porque no son útiles en este mapa
ax.set_xticks([])
ax.set_yticks([])

ax.set_title(
    "1. BUENO: Visualiza claramente el Happiness Score",
    fontsize=14,
    weight="bold",
    color="green",
)
plt.savefig("images/map_good.png", dpi=600)
plt.show()

# ===========================
# 2. GRÁFICO FEO
# ===========================
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
gdf.plot(ax=ax, color="lightgrey")
gdf.plot(ax=ax, column="Score", cmap="jet", legend=True, edgecolor="black", linewidth=2)

cbar = ax.get_figure().get_axes()[-1]
cbar.set_ylabel("Happiness Score", fontsize=12)

ax.set_title(
    "2. FEO: estilos que dificultan la lectura",
    fontsize=14,
    weight="bold",
    color="orange",
)
plt.savefig("images/map_ugly.png", dpi=600)
plt.show()

# ===========================
# 3. GRÁFICO MALO
# ===========================
gdf_proj = gdf.to_crs("ESRI:54030")
centroids = gdf_proj.geometry.centroid
gdf_points = gpd.GeoDataFrame(gdf_proj[["Score"]], geometry=centroids, crs=gdf_proj.crs)

fig, ax = plt.subplots(1, 1, figsize=(12, 6))
gdf_proj.plot(ax=ax, color="lightgrey")
mappable = gdf_proj.plot(
    ax=ax,
    column="Score",
    cmap="coolwarm",
    legend=True,
)
gdf_points.plot(ax=ax, markersize=gdf_points["Score"] * 10, alpha=0.5, color="blue")

cbar_ax = ax.get_figure().get_axes()[-1]
cbar_ax.set_ylabel("Happiness Score")

ax.set_title(
    "3. MALO: color y tamaño representan lo mismo",
    fontsize=14,
    weight="bold",
    color="red",
)
plt.savefig("images/map_bad.png", dpi=600)
plt.show()

# ===========================
# 4. GRÁFICO INCORRECTO
# ===========================
gdf_incorrect = gdf.copy()
gdf_incorrect["factor"] = (
    gdf_incorrect["Score"].fillna(0) / gdf_incorrect["Score"].max()
) * 3

gdf_incorrect["geometry"] = [
    scale(geom, xfact=f if f > 0 else 1, yfact=f if f > 0 else 1, origin="center")
    if geom is not None
    else None
    for geom, f in zip(gdf_incorrect["geometry"], gdf_incorrect["factor"])
]

fig, ax = plt.subplots(1, 1, figsize=(12, 6))
gdf_incorrect.plot(ax=ax, color="lightblue", edgecolor="black")
ax.set_title(
    "4. INCORRECTO: cartograma (demasiado) distorsionado según el Happiness Score",
    fontsize=14,
    weight="bold",
    color="darkred",
)
plt.savefig("images/map_incorrect.png", dpi=600)
plt.show()
```
